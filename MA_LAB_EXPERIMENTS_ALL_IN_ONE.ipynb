```python
num = int(input("Enter a number: "))
count = 0
for i in range(1, num + 1):
    if num % i == 0:
        count += 1
if count == 2:
    print("Prime number")
else:
    print("Not a prime number")
```

```python
num = int(input("Enter a number: "))
fact = 1
for i in range(1, num + 1):
    fact = fact * i
print("Factorial =", fact)
```

```python
num = int(input("Enter a number: "))
temp = num
rev = 0
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10
if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

X = np.array([1,2,3,4,5]).reshape(-1,1)
Y = np.array([2,4,5,4,5])

model = LinearRegression()
model.fit(X,Y)

m = model.coef_[0]
b = model.intercept_

print("Regression Line: y =", m, "* x +", b)

Y_pred = model.predict(X)

MSE = mean_squared_error(Y, Y_pred)
MAE = mean_absolute_error(Y, Y_pred)
RMSE = np.sqrt(MSE)
R2 = r2_score(Y, Y_pred)

n = len(Y)
p = 1

Adj_R2 = 1 - (1-R2)*(n-1)/(n-p-1)
SEE = np.sqrt(np.sum((Y-Y_pred)**2)/(n-2))

print("MSE =", MSE)
print("MAE =", MAE)
print("RMSE =", RMSE)
print("R2 =", R2)
print("Adjusted R2 =", Adj_R2)
print("SEE =", SEE)

plt.scatter(X,Y,label="Actual Data")
plt.plot(X,Y_pred,label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X=np.array([1,2,3]).reshape(-1,1)
Y=np.array([[2,3],[4,5],[6,7]])

model=LinearRegression()
model.fit(X,Y)

pred=model.predict(X)

print("Predicted Y1 and Y2:")
print(pred)

x_line=np.linspace(1,3,100).reshape(-1,1)
y_line=model.predict(x_line)

plt.figure(figsize=(10,6))
plt.scatter(X,Y[:,0],s=120)
plt.scatter(X,Y[:,1],s=120)
plt.plot(x_line,y_line[:,0],linewidth=3)
plt.plot(x_line,y_line[:,1],linewidth=3)
plt.title("Multivariate Linear Regression Prediction",fontsize=16)
plt.xlabel("X",fontsize=13)
plt.ylabel("Predicted Values",fontsize=13)
plt.grid(True,alpha=0.3)
plt.show()
```

```python
weather = ["Sunny","Sunny","Rainy","Rainy"]
play = ["Yes","Yes","No","No"]

p_yes = play.count("Yes")/len(play)
p_no  = 1 - p_yes

p_sy = sum(1 for w,p in zip(weather,play) if w=="Sunny" and p=="Yes")/play.count("Yes")
p_sn = sum(1 for w,p in zip(weather,play) if w=="Sunny" and p=="No")/play.count("No")
p_s  = weather.count("Sunny")/len(weather)

py_s = (p_sy*p_yes)/p_s
pn_s = (p_sn*p_no)/p_s

print(py_s, pn_s)
print("Play" if py_s>pn_s else "Don't Play")
```

Q1

```python
import numpy as np

X = np.array([[2,3],[3,4],[6,6],[7,7]])
c = np.array([[2,3],[6,6]])

while True:
    d = np.linalg.norm(X[:,None] - c, axis=2)
    g = np.argmin(d, axis=1)
    new_c = np.array([X[g==i].mean(0) for i in range(2)])
    if np.all(c == new_c): break
    c = new_c

print("Clusters:", [X[g==i] for i in range(2)])
print("Centroids:", c)
```

```python
import numpy as np

x = np.array([2,4,10,12,3,20,30,11,25])
c = np.array([2,20])

while True:
    g = np.argmin(abs(x[:,None]-c),1)
    nc = np.array([x[g==i].mean() for i in range(2)])
    if np.all(c==nc): break
    c = nc

print([x[g==i] for i in range(2)], c)
```
