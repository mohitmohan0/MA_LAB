```python
num = int(input("Enter a number: "))
temp = num
rev = 0
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10
if temp == rev:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

X = np.array([1,2,3,4,5]).reshape(-1,1)
Y = np.array([2,4,5,4,5])

model = LinearRegression()
model.fit(X,Y)

m = model.coef_[0]
b = model.intercept_

print("Regression Line: y =", m, "* x +", b)

Y_pred = model.predict(X)

MSE = mean_squared_error(Y, Y_pred)
MAE = mean_absolute_error(Y, Y_pred)
RMSE = np.sqrt(MSE)
R2 = r2_score(Y, Y_pred)

n = len(Y)
p = 1

Adj_R2 = 1 - (1-R2)*(n-1)/(n-p-1)
SEE = np.sqrt(np.sum((Y-Y_pred)**2)/(n-2))

print("MSE =", MSE)
print("MAE =", MAE)
print("RMSE =", RMSE)
print("R2 =", R2)
print("Adjusted R2 =", Adj_R2)
print("SEE =", SEE)

plt.scatter(X,Y,label="Actual Data")
plt.plot(X,Y_pred,label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression")
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X=np.array([1,2,3]).reshape(-1,1)
Y=np.array([[2,3],[4,5],[6,7]])

model=LinearRegression()
model.fit(X,Y)

pred=model.predict(X)

print("Predicted Y1 and Y2:")
print(pred)

x_line=np.linspace(1,3,100).reshape(-1,1)
y_line=model.predict(x_line)

plt.figure(figsize=(10,6))
plt.scatter(X,Y[:,0],s=120)
plt.scatter(X,Y[:,1],s=120)
plt.plot(x_line,y_line[:,0],linewidth=3)
plt.plot(x_line,y_line[:,1],linewidth=3)
plt.title("Multivariate Linear Regression Prediction",fontsize=16)
plt.xlabel("X",fontsize=13)
plt.ylabel("Predicted Values",fontsize=13)
plt.grid(True,alpha=0.3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X=np.array([[1,1],[2,1],[2,3],[3,3]])
y=np.array([1,1,-1,-1])

w=np.array([1,-1])
b=0

print("W1 =",w[0])
print("W2 =",w[1])
print("b =",b)

print("Constraint Values Yi(W.X+b)")
for i in range(len(X)):
    print(y[i]*(np.dot(w,X[i])+b))

x=np.linspace(0,4,100)
decision=(x-b)/1
margin1=(x-b-1)/1
margin2=(x-b+1)/1

plt.figure(figsize=(8,6))
plt.scatter(X[y==1][:,0],X[y==1][:,1],s=200,marker='o',label="+1 Class")
plt.scatter(X[y==-1][:,0],X[y==-1][:,1],s=200,marker='s',label="-1 Class")
plt.scatter(X[1][0],X[1][1],s=400,edgecolors='black',facecolors='none')
plt.scatter(X[2][0],X[2][1],s=400,edgecolors='black',facecolors='none')
plt.plot(x,decision,linewidth=3,label="Decision Boundary")
plt.plot(x,margin1,'--',linewidth=2,label="Margin")
plt.plot(x,margin2,'--',linewidth=2)
plt.title("Hard Margin Support Vector Machine",fontsize=16)
plt.xlabel("X1",fontsize=13)
plt.ylabel("X2",fontsize=13)
plt.grid(True,alpha=0.3)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X=np.array([2,4,0,1])
y=np.array([1,1,-1,-1])

w=2
b=-3

print("W =",w)
print("b =",b)

print("Constraint Values Yi(WXi+b)")
for i in range(len(X)):
    print(y[i]*(w*X[i]+b))

print("Support Vectors:",1,2)

x=np.linspace(-1,5,200)
decision=( -b )/w
margin1=(1-b)/w
margin2=(-1-b)/w

plt.figure(figsize=(10,5))
plt.scatter(X[y==1],np.ones_like(X[y==1]),s=250,marker='o',label="+1 Class")
plt.scatter(X[y==-1],-np.ones_like(X[y==-1]),s=250,marker='s',label="-1 Class")
plt.axvline(decision,linewidth=3,label="Decision Boundary")
plt.axvline(margin1,linestyle="--",linewidth=2,label="Margin")
plt.axvline(margin2,linestyle="--",linewidth=2)
plt.scatter(2,1,s=400,edgecolors='black',facecolors='none')
plt.scatter(1,-1,s=400,edgecolors='black',facecolors='none')
plt.title("1D Hard Margin Support Vector Machine",fontsize=16)
plt.xlabel("X",fontsize=13)
plt.yticks([-1,1])
plt.grid(True,alpha=0.3)
plt.legend()
plt.show()
```

```python
import numpy as np

X=np.array([[0,-1,1],[0,2,-1],[-1,0,2]])
y=np.array([1,-1,-1])
a=np.array([0.2,0.8,0.4])
b=0

w=np.sum((a*y)[:,None]*X,axis=0)

print("Weight Vector:",w)

for i in range(len(X)):
    f=np.sum(a*y*np.dot(X,X[i]))+b
    c=1 if f>0 else -1
    print("X",i+1,"=",X[i],"f(x)=",f,"Predicted Class =",c)
```
